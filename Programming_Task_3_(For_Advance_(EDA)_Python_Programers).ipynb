```python
#1 question
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        maxx= 0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                prod = (nums[i]-1)*(nums[j]-1)
                maxx = max(prod , maxx)
        return maxx
```

```python
#2 question
def numTeams(rating):
    res = 0
    n = len(rating)
    for item in range(n - 2):
        for item1 in range(item + 1, n - 1):
            for item2 in range(item1 + 1, n):
                if (rating[item] < rating[item1] < rating[item2]) or (rating[item] > rating[item1] > rating[item2]):
                    res += 1
    return res
rating = [2,5,3,4,1]
print(numTeams(rating))
```

```python
#3 question
def busyStudent(startTime, endTime, queryTime):
    count = 0
    for item in range(len(startTime)):
        if startTime[item] <= queryTime  <=endTime[item]:
            count += 1
    return count
startTime = [1,2,3]
endTime = [3,2,7]
queryTime = 4
print(busyStudent(startTime, endTime, queryTime))
```

```python
#4 question
def numberOfSteps(num):
        count = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
                count += 1
            else:
                num -= 1
                count += 1
        return count
num = 14
print(numberOfSteps(num))
```

```python
#5 question
def countBits(n):
    sum=0
    list1 = []
    for item in range(n+1):
        sum=bin(item).count("1")
        list1.append(sum)
    return list1
n=2
print(countBits(n))
```
