```python
#question 1
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        sum = 0
        res = []
        for i in nums:
            sum += i
            res.append(sum)
        return res
```

```python
#question 2
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        temp=[]
        for i in range(n):
            temp.append(nums[i])
            temp.append(nums[i+n])

        return temp
```

```python
#question 3
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        res=[]
        for i in candies:
            if (extraCandies+ i >= max(candies)):
                res.append(True)
            else:
                res.append(False)
        return res
```
