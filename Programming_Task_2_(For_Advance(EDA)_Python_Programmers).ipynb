```python
# 1st question
class Solution:
    def defangIPaddr(self, address: str) -> str:
        res = ""
        for i in address:
            if (i=='.'):
                res += '[.]'
            else:
                res += i
        return res
```

```python
# 2nd question
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        ct = 0
        for num in nums :
            no = 0
            while(num > 0):
                num = num // 10
                no += 1
            if(no % 2 == 0):
                ct +=1

        return ct
```

```python
# 3rd question
class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        ct =0
        for i in range(len(nums)):
            for j in range(i+1 , len(nums)):
                if(nums[i] == nums [j] ):
                    ct +=1

        return ct
```

```python
# 4th question
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        sort_nums = sorted(nums)
        res = []
        box = {}
        for i in range(len(nums)):
            if sort_nums[i] not in box :
                box[sort_nums[i]] = i

        for num in nums:
            res.append(box[num])

        return res
```

```python
# 5th question
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        sum = 0
        product = 1
        while(n>0 ):
            dig = n % 10
            sum += dig
            product *= dig
            n= n // 10
        return product - sum
```

```python
# 6st question
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        arr= []
        ans = 0
        for i in range(n):
            arr.append(start + 2*i)
        for j in arr:
            ans = ans ^ j

        return ans
```
